This notebook exports subsets of the full nasdaq_external_data.csv file downloaded from https://huggingface.co/datasets/Zihan1004/FNSPID/tree/main/Stock_news. 

This is a comprehensive financial news dataset curated by authors of the paper https://arxiv.org/html/2402.06698v1#S3.

# Imports

```python
import pandas as pd
```

```python
%%time
df1 = pd.read_csv(r"C:\Users\mushj\Downloads\RAW FINANCE DATA\nasdaq_exteral_data.csv")
```

# Summary 

```python
df1['Stock_symbol'].nunique()
```

```python
# print first row in detail
for i,j in df1.iloc[0].items():
    print(i)
    print('\t', j, '\n')
```

```python
# compute date ranges for each stock
df_date_ranges = df1.groupby('Stock_symbol').agg({'Date': [min, max]})
df_date_ranges['range'] = pd.to_datetime(df_date_ranges[('Date','max')]) - pd.to_datetime(df_date_ranges[('Date','min')])
```

```python
# summarize date ranges
df_date_ranges['range'].describe()
```

```python
# show longest date ranges
df_date_ranges.sort_values('range').tail(50)
```

# View specific ticker

```python
# select ticker
ticker = 'CRWD'

# get subset of data related to ticker
sub_df = (
    df1
    .query("Stock_symbol == @ticker")
    .drop('Unnamed: 0', axis=1)
)

print('Missing values summary:')
display(sub_df.isna().mean())

print('\nStart and end dates:')
print(sub_df.Date.min(), ' - ', sub_df.Date.max())
```

# Export subsets

```python
output_path = "C:/Users/mushj/Downloads/RAW FINANCE DATA/"

symbols = [
    'TSLA',
    'MSFT',
    'AMZN',
    'AAPL',
    'NVDA',
    'CRWD',
]
```

```python
for export_ticker in symbols:
    out_df = (
        df1
        .query("Stock_symbol == @export_ticker")
        .drop('Unnamed: 0', axis=1)
    )
    out_df.to_csv(output_path+f"FNSPID_{export_ticker}.csv", index=False)
    print('Exported --', export_ticker, out_df.shape[0])
```
